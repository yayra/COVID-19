# **Data Cleaning in python (Pandas).**

Dataset was downloaded from [Kaggle](https://www.kaggle.com/datasets/kunwardeepak/covid19-infected-person-list). The original dataset consists of 2,676,311 rows and 33 columns.


```python
import pandas as pd # importing libraries
import numpy as np
import datetime as dt
```

```python
!git clone https://github.com/yayra/Covid-19 
!unzip "/content/Covid-19/archive.zip"       # clone zip file from my github and unzip 
df=pd.read_csv("latestdata.csv")
df.head()
```

#Data integrity check
Checking for missing values (NaN) and drop columns, which will not be used in the next step of data visualization.

```python
df.isnull().sum()/len(df)*100   # check the percentage of missing values in every column
```

```python
df.drop(['city','geo_resolution', 'lives_in_Wuhan', 
         'travel_history_location', 'reported_market_exposure',
         'additional_information', 'chronic_disease',
          'source', 'sequence_available', 'date_death_or_discharge', 
         'notes_for_discussion', 'location', 
         'admin3', 'admin2', 'admin1', 'admin_id', 'country_new',
         'data_moderator_initials'], axis=1, inplace=True) #dropping columns with highest percentage of missing values and not to be used in further analysis
```

```python
df.info() #Check datatypes of the values
```

#Cleaning column 'age'.


```python
df['age'].unique() #checking 'age' column on outliers
```

```python
df['age'] = df['age'].str.replace(r'(^.*mon.*$)', '1').str.replace(r'(^.*week.*$)', '1').str.replace('+', '') # replacing values containing 'week', 'month' substrings with value equal to 1. 
df[['age_lower','age_upper']] = df['age'].str.split('-', expand=True) #Split age column into 2 columns: 'age_lower','age_upper'. Only the lower boundary of the age range (eg in an age range 60-70 lower part 60 will be left )
df.drop(['age','age_upper'], axis=1, inplace=True) # Drop columns 'age' and 'age upper'.
df.rename(columns = {'age_lower':'age'}, inplace=True) #renaming age_lower column name into 'age'.
df['age'].fillna(0, inplace=True) #Replacing missing values with 0. 
df['age'] = np.round(df['age'].astype(float)) #Rounding up and converting object to float. 
df['age'] = df['age'].astype(int) #Converting float to integer. 
```

#Cleaning columns 'date_onset_symptoms','date_admission_hospital', 'date_confirmation'. 	

```python
df[df['date_onset_symptoms'].str.len()>10] #Identify outliers
```

```python
df['date_onset_symptoms']= df['date_onset_symptoms'].replace({'10.01.2020 - 22.01.2020':'16.01.2020','01.01.2020-12.01.2020':'06.01.2020'}).str.strip() #Replace and strip values
df['date_admission_hospital']=df['date_admission_hospital'].replace({'18.01.2020 - 23.01.2020':'21.01.2020'}).str.strip()
df[['date_confirmed', 'fordrop']] = df['date_confirmation'].str.split('-', expand=True) #Splitting 'date_confirmation' column into 2 columns
df=df.drop(['date_confirmation', 'fordrop'], axis=1) #Drop unnecessary columns
df[['date_onset_symptoms', 'date_admission_hospital', 'date_confirmed']] = df[['date_onset_symptoms', 'date_admission_hospital', 'date_confirmed']].apply(pd.to_datetime) #convert the datatype from object to datetime
```

#Cleaning column 'outcome'

```python
df['outcome']=df['outcome'].str.lower() 
df['outcome']=df['outcome'].fillna(0)
df['outcome'].value_counts()
```

All values in 'outcome' column will be divided into 4 categories:
1. Hospitalized
2. Dead
3. Not hospitalized
4. Undefined




```python
Hospitalized = ['critical condition, intubated as of 14.02.2020', 
       'discharge', 'discharged', 'discharged from hospital',
       'recovered', 'severe', 'stable', 
       'symptoms only improved with cough. currently hospitalized for follow-up.',
       'treated in an intensive care unit (14.02.2020)', 'alive',
       'stable condition', 'under treatment',
       'critical condition', 'receiving treatment', 'severe illness',
       'unstable', 'hospitalized']
Dead = ['death', 'died', 'dead', 'deceased']
Not_hospitalized = ['not hospitalized', 'released from quarantine', 'recovering at home 03.03.2020']
Other = ['0', 'migrated', 'migrated_other','https://www.mspbs.gov.py/covid-19.php']

df['outcome'] = df['outcome'].apply(lambda x: 'Hospitalized' if x in Hospitalized else x )
df['outcome'] = df['outcome'].apply(lambda x: 'Dead' if x in Dead else x )
df['outcome'] = df['outcome'].apply(lambda x: 'Not hospitalized' if x in Not_hospitalized else x)
df['outcome'] = df['outcome'].apply(lambda x: 'Undefined' if x in Other else x )
    
```
